```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('loss.log', 'r') as f:
    lines = [{'loss': line.strip()} for line in f.readlines()]
```

```python
df = pd.DataFrame(lines, dtype=float)
```

```python
df.describe()
```

```python
df['min'] = df.loss.min()
print('Min Loss: {} Index: {}'.format(df.loss.min(), df.loss.idxmin()))
df.plot(figsize=(20, 10))
```
